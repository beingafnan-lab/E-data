```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv" , encoding = 'latin1')
```

```python
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day Of  Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month , x = 'Order Month' , y = 'Sales' , title = 'Monthly Sales Analysis')
fig.show()
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
fig = px.pie(sales_by_category, values = 'Sales' , names = 'Category' , hole = 0.4 , color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_traces(textposition = 'inside' , textinfo = 'percent+label')
fig.update_layout(title = 'Sales Analysis by Category' , titlefont = dict(size = 24))
fig.show()
```

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory , x = 'Sub-Category' , y = 'Sales' , title = 'Sales Analysis by Subcategory')
fig.show()
```

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig = px.pie(profit_by_category , values = 'Profit' , names = 'Category' , hole = 0 , color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_traces(textposition = 'inside' , textinfo = 'percent+label')
fig.update_layout(title = 'Profit Analysis by Category' , titlefont = dict(size = 24))
fig.show()
```

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
fig = px.bar(profit_by_subcategory , x = 'Sub-Category' , y = 'Profit' , title = 'Profit Analysis')
fig.show()
```

```python
sales_to_profit_ratio = data['Sales'].sum()/data['Profit'].sum()
sales_to_profit_ratio
```

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales' : 'sum' , 'Profit' : 'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x = sales_profit_by_segment['Segment'],
                     y = sales_profit_by_segment['Sales'],
                     name = 'Sales',
                     marker_color = color_palette[9]))
fig.add_trace(go.Bar(x = sales_profit_by_segment['Segment'],
                     y = sales_profit_by_segment['Profit'],
                     name = 'Sales',
                     marker_color = color_palette[8]))
fig.update_layout(title = 'Sales and Profit Analysis by Customer Segment' , 
                  xaxis_title = 'Customer Segment',
                  yaxis_title = 'Amount')
fig.show()
```
